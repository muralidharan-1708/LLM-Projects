```python
question_answerer = pipeline("question-answering")

# Define the context
context = '''
Python is a high-level, general-purpose programming language. Its design philosophy emphasizes code readability with the use of significant indentation.
Python is dynamically type-checked and garbage-collected. It supports multiple programming paradigms, including structured (particularly procedural),
object-oriented, and functional programming. It is often described as a "batteries included" language due to its comprehensive standard library.

Guido van Rossum began working on Python in the late 1980s as a successor to the ABC programming language, and he first released it in 1991 as Python 0.9.0.
Python 2.0 was released in 2000. Python 3.0, released in 2008, was a major revision not completely backward-compatible with earlier versions.
Python 2.7.18, released in 2020, was the last release of Python 2.

Python consistently ranks as one of the most popular programming languages, and it has gained widespread use in the machine learning community.
'''

# Define multiple questions
questions = [
    "What is Python?",
    "Who developed Python?",
    "In which year was Python first released?"
]

# Loop through each question and get an answer
for q in questions:
    result = question_answerer(question=q, context=context)
    print(f"Q: {q}")
    print(f"A: {result['answer']}\n")
```
